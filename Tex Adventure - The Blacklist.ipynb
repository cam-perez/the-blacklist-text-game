```python
# Stating game purpose
"""A brief description of the code

Created by Camila Perez as part of the course Data Science: Python - DAT-5301

This Text Adventure Game is themed on the first two seasons of the TV series
The Blacklist (2013 - present), created by Jon Bokenkamp. It is not necessary
to have watched The Blacklist previously to enjoy or understand the game
because the player will be provided with the necessary information to complete
each one of the three total stages. 

Errors/Bugs identified as of October 25, 2021: none identified"""

# Importing necessary packages
import random

#Creating message list for unexpected entry 
invalid_entry = ["Invalid entry. Please try again.", "Access denied. Try again",
                "Clearance access denied. Please try again.",
                "Try again please."]

#Creating message list printed before fail function
exit_message = ["Goodbye, Agent", "Game Over", "End",
                "Reddington's story ends here for now"]

#Creating message list for fail function
fail_message = [f"""
Agent, this path runned you out of time, the Cabal has discovered Reddington's
location and his death is confirmed.

The task force no longer exist.""",f"""
Agent, you were unable to solve the case and Reddington decease has been confirmed.

The task force no longer exist."""]

#Creating list stuffed animal options
stuffed_animal = ["bunny", "giraffe", "lion", "cat", "elephant", "tiger"]

# Defining function
def registration_form():
    """ This function is meant to collect the player's codename, provide a
    context for the game and eventually start the game."""

    # Requesting user to provide a name to start the game.
    codename = input(prompt = f"""
Welcome to the Post Office.

Please enter your codename.\n""").title()
    
    # Printing stage information
    print(f"""
file__red_01

Agent {codename}, you are now part of the FBI task force called The Post Office.
Your partner is Agen Keen, who works closely with Raymond Reddington, one of
the 10 FBI most wanted criminals in the FBI list.

Reddington suddenly surrendered a few months ago and became an informant as 
part of his plan to fight the Cabal, a shadow government constantly threatinng
his life without taking any action yet as Reddington has affirmed for several
years to be in possession of The Fulcrum.

One hour ago the Cabal's Director realesed an assassination order against
Reddington, he was shot and is in surgery as you read this message. Your
mission is to Collect The Fulcrum and descripher it to block the Director's order.\n
""")
    
    # Defining while loop
    while True:
        
        # Colecting and converting input into lowercase
        join_mission_decision = input(prompt = "Do you accept the mission?\n").lower()

        # If condition leads to stage 1
        if 'y' in join_mission_decision or 'sure' in join_mission_decision or 'ok' in join_mission_decision:
            print(f"""Agent {codename}, the task force has found new leads.\n""")
            mission_1() # Moving player to next stage
            break

        # Elif condition to quit the game
        elif 'n' in join_mission_decision:
            print(f"""Agent {codename}, thank you for your time.\n""")
            fail() # Exiting the game
            break

        # Else condition for unexpected entry.
        else:
            print(f"""{random.choice(invalid_entry)} \n""")
        
# Defining stage 1        
def mission_1():

    # Printing stage information
    print(f"""
loading_file_red_0X23X

Agen Keen found a suitcase and a key in Reddingston's apartment, but no one in
the task force knows what this device is or what it could do.

end_ of_file_red_0X23X.\n
""")

    # Defining while loop
    while True:
        
        # Colecting and converting input into lowercase
        box = input(prompt = "Do you want to investigate further or drop the lead?\n").lower()

        # If condition leads to stage 1
        if 'investigate' in box or 'follow' in box or 'continue' in box or 'y' in box:            
            print(f"""
Agent, how have discovered an invaluable lead for the case.
Now figure out how to make this device work'.\n
""")
            mission_2() # Moving player to next stage
            break
        
        # Elif condition to quit the game
        elif 'drop' in box or 'skip' in box or 'abandon' in box or 'n' in box:
            fail() # Exiting the game
            break

        # Else condition for unexpected entry.
        else:
            print(f"""{random.choice(invalid_entry)} \n""")


# Defining stage 2
def mission_2():
    
    # Printing stage information
    print(f"""
loading_file_red_0XX

During the last year Agent Keen has been collecting evidence related to
Reddington and his interest on her as she believes that he could be her father.

end_ of_file_red_0XX


A slightly burned stuffed animal that Agent Keen had as a child is part of the
evidence.\n
""")

    # Defining while loop
    while True:
        
        # Colecting and converting input into lowercase
        bunny_inv = input(prompt = "Agent, do you want to investigate or drop this lead?\n").lower()

        # If condition first indent level
        if 'investigate' in bunny_inv or 'follow' in bunny_inv or 'continue' in bunny_inv or 'y' in bunny_inv:
            
            # Printing sub-stage information
            print(f"""
During your analysis you notice the stuffed {random.choice(stuffed_animal)}'s neck has been sewed again,
this could be due to wear or time effect.

Agent, decide if you want to open the bunny to investigate the filling material
or move on.\n
""")
            
            # Colecting and converting input into lowercase
            fill_bunny = input(prompt = "Type your decision. \n").lower()
            
            # If condition second indent level
            if 'no' in fill_bunny or 'move on' in fill_bunny or 'abandon' in fill_bunny or 'drop' in fill_bunny:
                print(f"""
{random.choice(fail_message)}\n""")
                fail() # Exiting the game
                break
            
            # Elif condition second indent level
            elif 'cut' in fill_bunny or 'open' in fill_bunny or 'investigate' in fill_bunny or 'yes' in fill_bunny:
                print(f"""
Agent, you have discovered a device that resembles a chip and Reddington's
associates have a new lead: 

<Find Leonard Caul> \n
""")
                mission_3() # Moving player to next stage
                break
            
            # Ilse condition second indent level
            else:
                print(f"""{random.choice(invalid_entry)}\n""")
                break
        
        # Elif condition first indent level
        elif 'n' in bunny_inv or 'drop' in bunny_inv or 'move on' in bunny_inv:
            print(f"""
{random.choice(fail_message)}\n
""")
            fail() # Exiting the game
            break
        
        # Else condition first indent level for unexpected entry.
        else:
            print(f"""
{random.choice(invalid_entry)}\n""")

# Defining stage 3
def mission_3():
    
    # Printing stage information
    print(f"""
While looking for Leonard Caul, you found a phone number in Reddingston's
appartment with L typed at the bototm of the page. 

The last number is missing.\n
""")

    # Defining while loop
    while True:
        
        # Colecting and converting input into lowercase
        caulcall = input(prompt = "Agent, do you want to try to guess and call the number?\n").lower()

        # If condition leads to stage next stage
        if 'investigate' in caulcall or 'y' in caulcall or 'ok' in caulcall or 'sure' in caulcall:
            
            # Declaring objects
            missing_number = random.randint(1,9) # random number static
            attempt = 4

            # Looping until player is out of guesses
            while attempt > 0:
                attempt = input(prompt = "Agent, try guessing the missing number here.\n")

                # Converting player guess for numerical comparison
                attempt = int(missing_number)

                # If condition, first loop indentation
                if attempt == missing_number:
                    print(f"""
Mr. Caul himself picked the call, you traced the location and met with him.

Congratulations, Agent, you have completed your first mission.\n
""")
                    break
                
                # Elif condition, first loop indentation
                elif attempt != missing_number:
                    attempt -= 1 # Decreasing remaining attempts

                    # Printing message for remaining attempts
                    if attempt > 0:
                        print(f"""
Wrong number, agent. You have {attempt} guesses remaining.\n""")

                    elif attempt == 0:
                        print(f"""
Agent, there is no remaining attempts.
The correct number was {missing_number}.\n""")
                        fail()

                    # Else condition, second loop indentation
                    else:
                        print("Bug in incorrect missing number.")

                # Else condition, first loop indentation
                else:
                    print("Agent, try with a number in numerical form, e.i. 1, 2, etc.")               
            break #Breaking loop from line 214
        
        # Elif condition leads to fail
        elif 'n' in caulcall or 'skip' in caulcall or "shouldn't" in caulcall:
            print(f"""
{random.choice(fail_message)}\n""")
            fail()
            break
        
        #Else conditional statement for unexpected values
        else:
            print(f"""{random.choice(invalid_entry)} \n""")
          
            
# Defining fail function
def fail():
    print("Your access to the Post Office has been removed.\n")
    fail_reply = input(prompt = f"""
Press any key to exit the facility or
type yes to register and play again.\n""")
    if 'y' in fail_reply:
        registration_form()
        
    else:
        print(f"""{random.choice(exit_message)}""")

registration_form()
```
